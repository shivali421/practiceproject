```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Identify categorical columns
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome','subscribed']
```

```python
# Apply LabelEncoder to all categorical columns
label_encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
# Splitting into features (X) and target (y)
X = df.drop('subscribed', axis=1)
y = df['subscribed']
```

```python
# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Explore class distribution of the target variable
sns.countplot(x='subscribed', data=df)
plt.title('Class Distribution of subscribed')
plt.xlabel('subscribed')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze distribution of other features in relation to the target
numeric_features = ['ID','age', 'balance', 'day','duration','campaign','pdays', 'previous']
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
```

```python
# Numeric features vs. Target
for feature in numeric_features:
    sns.boxplot(x='subscribed', y=feature, data=df)
    plt.title(f'{feature} vs subscribed')
    plt.show()
```

```python
# Categorical features vs. Target
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='poutcome', data=df)
    plt.title(f'{feature} vs poutcome')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='poutcome', labels=['No', 'Yes'])
    plt.show()
```

```python
# Extract relevant information from "month" and "day_of_week"
# Convert "month" to numerical values
month_mapping = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}
```

```python
df['month_num'] = df['month'].map(month_mapping)
```

```python
# Create new features
# Interaction feature: duration * campaign (assumes a combined effect)
df['duration_campaign_interaction'] = df['duration'] * df['campaign']
```

```python
# Indicator feature: whether the previous marketing campaign was successful
df['previous_success'] = np.where(df['previous'] > 0, 1, 0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
# Define models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Support Vector Machine': SVC()
}
```

```python
# Training and evaluating models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    
    accuracy = accuracy_score(y_val, y_pred)
    print(f'{name} Accuracy: {accuracy:.2f}')
    
    # For multiclass classification, use 'macro' average for precision, recall, and F1-score
    precision = precision_score(y_val, y_pred, average='macro')
    recall = recall_score(y_val, y_pred, average='macro')
    f1 = f1_score(y_val, y_pred, average='macro')
    
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1-Score: {f1:.2f}\n')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
# Define the model (for example, Random Forest) for hyperparameter tuning
model = RandomForestClassifier()
```

```python
# Define hyperparameters and their possible values for grid search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Define the scoring metric (F1-score)
scorer = make_scorer(f1_score)
```

```python
# Perform grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, scoring=scorer, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Print the best hyperparameters
print("Best Hyperparameters:")
print(grid_search.best_params_)
```

```python
# Define the model (for example, Random Forest) for hyperparameter tuning
model = RandomForestClassifier(n_estimators=100, max_depth=10)  # You can use the best hyperparameters from earlier
model.fit(X_train, y_train)
```

```python
# Predict on the validation set
y_pred = model.predict(X_val)
```

```python
# Calculate the F1-score for multiclass classification
f1 = f1_score(y_val, y_pred, average='weighted')  # Use 'weighted' average
print(f"Validation F1-Score: {f1:.2f}")
```

```python
# Evaluate the best model on the validation set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val)
f1 = f1_score(y_val, y_pred)
print(f"Validation F1-Score: {f1:.2f}")
```
